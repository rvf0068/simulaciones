Esta es una simulación de la paradoja del cumpleaños.



```python
from random import choice

días = list(range(1, 366))

def fechas(personas):
    cumple = []
    for i in range(personas):
        día = choice(días)
        cumple.append(día)
    return cumple
```

```python
todas = fechas(25)
todas
```

```python
def repetidos(lista):
    for i in range(len(lista)):
        for j in range(i+1, len(lista)):
            if lista[i] == lista[j]:
                return (lista[i], lista[j])
    else:
        return False
```

```python
repetidos(todas)
```

```python
def simulaciones(n, personas):
    repeticiones = 0
    for i in range(n):
        lista = fechas(personas)
        if repetidos(lista):
            repeticiones = repeticiones+1
    return repeticiones
```

```python
simulaciones(500, 22)
```
